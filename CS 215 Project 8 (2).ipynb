```python
# Importing necessary packages

import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
# Opening the HTML file and converting it to soup

html = open(r"C:\Users\colby\Downloads\colby._ramsey_20230413\likes\liked_comments.html", encoding = 'utf8')
soup = BeautifulSoup(html, 'html.parser')
print(html)
```

```python
# Finding all individal likes

likes = soup.find_all('div',{"class":"_a6-p"})
```

```python
# Printing to make sure everything went smoothly

likes
```

```python
# Going through likes and finding them individually

for l in likes:
    print(l.text)
```

```python
# Eliminating the emoji and finding the date and time of each like

for l in likes:
    print(l.text[1:])
```

```python
# Creating a table for the timestamps

timestamps = pd.Series(['timestamp'])
```

```python
timestamps
```

```python
# Adding the elements of likes to the timestamp table

for l in likes:
    row = pd.Series({'timestamp': l.text[1:]})
    timestamps = pd.concat([timestamps, row])
```

```python
timestamps
```

```python
# Downloading file for use by my partner

timestamps.to_csv("colby.csv")
```
