```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
colby_timestamps = pd.read_csv('colby.csv')
eli_timestamps = pd.read_csv('EliTimestamps.csv')
```

```python
colby_timestamps
```

```python
colby = pd.to_datetime(colby_timestamps.iloc[1:,1])
colby
```

```python
eli_timestamps
```

```python
eli = pd.to_datetime(eli_timestamps.iloc[:,1])
eli
```

```python
plt.title('Timestamps for Eli and Colby')
plt.gca().get_yaxis().set_visible(False)

plt.scatter(colby, np.random.random(len(colby)), color='blue', label='Colby', alpha=0.5)
plt.scatter(eli, np.random.random(len(eli)) + 2, color='red', label='Eli', alpha=0.5)
plt.legend()
plt.show()
```
